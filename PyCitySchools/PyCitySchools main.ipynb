```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data file and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
#school_data.head()
```

```python
# Combine the data into a single dataset.
#student_data.head()
```

```python
#merging school data and student data.
school_data_complete = pd.merge(student_data, school_data,how="left",on=["school_name","school_name"])
#
school_data_complete.head()
#calculate total school count.
Total_schools = school_data["school_name"].count()
#calculate total student count.
Total_Student = student_data["student_name"].count()
#calculate total budget sum.
Total_budget = school_data["budget"].sum()
#calcutalte average math score. 
Average_Math_Score = student_data["math_score"].mean()
#calculate average reading score.
Average_Reading_Score = student_data["reading_score"].mean()
# finding student who are pass in math.
student_data_maths_score_df = pd.DataFrame(school_data_complete.loc[student_data["math_score"]>=70,:])
# getting the count of student who pass the math score
pass_math_count = student_data_maths_score_df["student_name"].count()
#Percentage of students with passing math score
passing_math_pct = (pass_math_count/Total_Student)*100

#finding student who are pass the reading score.
student_data_reading_score_df = pd.DataFrame(school_data_complete.loc[student_data["reading_score"]>=70,:])
#getting the count of student who pass the reading score.
pass_reading_count = student_data_reading_score_df["student_name"].count()
#pecentage of students with passing english score.
passing_reading_pct = (pass_reading_count/Total_Student)*100
# Overall passing rate
overall_pass_df = pd.DataFrame(student_data_maths_score_df.loc[student_data_maths_score_df["reading_score"]>=70,:])
#getting the count of overall .
overall_pass_count  = overall_pass_df["student_name"].count()

#prsentage of overall passing.
overall_pass_pct = (overall_pass_count/Total_Student)*100


#print(Total_schools)
#print(Total_Student)
#print(Total_budget)
#print(Average_Math_Score)
#print(Average_Reading_Score)
#print(math_score)
#print(passing_math_pct)
#print(pass_reading_count)
#print(passing_reading_pct)
#print(overall_pass_count)
#print(overall_pass_pct)      
```

```python
#creating a data frame 
district_summary = {"Total Schools": [Total_schools],"Total Students":[Total_Student],"Total Budget":[Total_budget],
                   "Average Math Score":[Average_Math_Score],"Average Reading Score":[Average_Reading_Score],
                    "% Passing Math":[passing_math_pct],"% Passing Reading":[passing_reading_pct],"% Overall Passing":[overall_pass_pct]}                   
```

```python
#printing data frame summary
district_summary_df = pd.DataFrame(district_summary)
district_summary_df['Total Students'] = district_summary_df.apply(lambda x: "{:,.0f}".format(x['Total Students']), axis=1)
district_summary_df['Total Budget'] = district_summary_df.apply(lambda x: "${:,.2f}".format(x['Total Budget']), axis=1)
district_summary_df
```

```python
#getting group total student count .
Total_student_groupby_df = school_data_complete[["school_name","type","student_name"]].groupby(["school_name","type"]).count().reset_index().rename(columns={'student_name':'Total Students','type':'School Type'})
#getting group total budget sum.
Total_budget_groupby_df = school_data[["school_name","budget"]].groupby(["school_name"]).sum().reset_index().rename(columns={'budget':'Total School Budget'})
#getting group average math score mean.
Average_math_score_groupby_df = school_data_complete[["school_name","math_score"]].groupby(["school_name"]).mean().reset_index().rename(columns={'math_score':'Average Math Score'})
#getting group reading score mean .
Average_reading_score_groupby_df = school_data_complete[["school_name","reading_score"]].groupby(["school_name"]).mean().reset_index().rename(columns={'reading_score':'Average Reading Score'})
#getting group passing math score count.
passing_math_score_groupby_df = student_data_maths_score_df[["school_name","student_name"]].groupby(["school_name"]).count().reset_index().rename(columns={'student_name':'pass_math'})
#getting group passing reading score count.
passing_reading_score_groupby_df = student_data_reading_score_df[["school_name","student_name"]].groupby(["school_name"]).count().reset_index().rename(columns={'student_name':'pass_reading'})
#getting group passing overall count. 
passing_overall_groupby_df = overall_pass_df[["school_name","student_name"]].groupby(["school_name"]).count().reset_index().rename(columns={'student_name':'pass_overall'})
```

```python
#merging all the data frame
merge_df_1 = pd.merge(Total_student_groupby_df,Total_budget_groupby_df ,how="inner",on="school_name")
merge_df_2 = pd.merge(merge_df_1,Average_math_score_groupby_df ,how="inner",on="school_name")
merge_df_3 = pd.merge(merge_df_2,Average_reading_score_groupby_df ,how="inner",on="school_name")
merge_df_4 = pd.merge(merge_df_3,passing_math_score_groupby_df ,how="inner",on="school_name")
merge_df_5 = pd.merge(merge_df_4,passing_reading_score_groupby_df ,how="inner",on="school_name")
merge_df_6 = pd.merge(merge_df_5,passing_overall_groupby_df ,how="inner",on="school_name")

#getting the per student budget.
merge_df_6["Per Student Budget"] = merge_df_6["Total School Budget"]/merge_df_6["Total Students"]
#calculate the passing percentage in math. 
merge_df_6["% Passing Math"] = (merge_df_6["pass_math"]/merge_df_6["Total Students"])*100
#calculate the passing percentage in reading.
merge_df_6["% Passing Reading"] = (merge_df_6["pass_reading"]/merge_df_6["Total Students"])*100
#calculate the overall passing percentage. 
merge_df_6["% Overall Passing"] = (merge_df_6["pass_overall"]/merge_df_6["Total Students"])*100
#droping the additinal columns.
columns = ['pass_math','pass_reading','pass_overall']
merge_df_6.drop(columns, inplace=True, axis=1)
#reordering the cloumn.
merge_df_6= merge_df_6[['school_name','School Type','Total Students','Total School Budget','Per Student Budget',
                        'Average Math Score','Average Reading Score','% Passing Math','% Passing Reading',
                        '% Overall Passing']]
#setting school name as index.
merge_df_6= merge_df_6.set_index('school_name')
#settinfg the index name to none.
merge_df_6.index.names = [None]
 
```
